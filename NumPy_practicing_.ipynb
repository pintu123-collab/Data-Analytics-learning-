#**STACK:-**
**a)vstack(vertical stack)**

**b)hstack(horizontal stack)**

**c)dstack(defth stack)**

**V-STACK**

```python
# code example
import numpy as np

a=np.array([[1,2],
            [3,4]])
b=np.array([[5,6],
            [7,8]])
np.vstack((a,b))
```

**H-STACK**

```python
# code example

import numpy as np

a=np.array([[1,2],
            [3,4]])
b=np.array([[5,6],
            [7,8]])
np.hstack((a,b))
```

**D-STACK**

```python
# code example

import numpy as np

a=np.array([[1,2],
            [3,4]])
b=np.array([[5,6],
            [7,8]])
np.dstack((a,b))
```
